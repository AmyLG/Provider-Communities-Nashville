## 2-Cleaning zipcodes from Nppes file

The zipcodes in the nppes dataset are either 5 or 9 digits. We clean the data to have 5 digits only and add 0 if less than 5 digits (zeros can be dropped when reading the data into a dataframe).

```python
db = sqlite3.connect('../data/nppes_lite.sqlite') #reopen the connection

query = """
SELECT * 
FROM nppes
"""
nppes = pd.read_sql(query,db)

db.close() 
```

```python
nppes.shape
```

```python
nppes.info()
```

```python
#nppes["zip"].value_counts()
```

```python
### float to object conversion for zip column
nppes['zip'] = nppes['zip'].astype(str)
```

```python
### Remove .0
nppes['zip'] = nppes['zip'].str.replace('\.0','')
```

```python
### Some zipcodes are less than 9, add 0 if it's the case
def zip_finder(zips):
    if len(zips)==9:
        return zips[:5]
    if len(zips)==8:
        return '0'+zips[:4]
    if len(zips)==7:
        return '00'+zips[:3]
    if len(zips)==5:
        return zips
    if len(zips)==4:
        return '0'+zips
    if len(zips)==3:
        return '00'+zips
    else:
        return 'no real zip'
```

```python
### Apply zip_finder function and keep only values with real zipcodes
nppes["zip"] = nppes.zip.apply(zip_finder)
nppes = nppes[nppes.zip!="no real zip"]
```

```python
nppes.head()
```

```python
#nppes["zip"].value_counts()
```

## 4-Merge Taxonomy to nppes

```python
db = sqlite3.connect('../data/nppes_lite.sqlite') #reopen the connection

query = """
SELECT code, grouping, classification, specialization 
FROM nucc_taxonomy
"""
taxonomy = pd.read_sql(query,db)

db.close() 
```

```python
taxonomy.head()
```

```python
# merge
nppes_cbsa_tax = nppes_cbsa.merge(taxonomy, left_on='taxonomy_code', right_on='code')
```

```python
nppes_cbsa_tax.shape
```

```python
# transform entity_type_code from float to integer
nppes_cbsa_tax['entity_type_code'] = nppes_cbsa_tax['entity_type_code'].astype(int)
```

```python
nppes_cbsa_tax.head()
```

## 5-Filter for Nashville cbsa only:

```python
#filter for Nashville CBSA code 34980
nash_nppes = nppes_cbsa_tax[nppes_cbsa_tax.cbsa == 34980]
```
